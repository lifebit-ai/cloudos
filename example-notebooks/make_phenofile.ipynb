Now we know the IDs of the phenotypes we want to include in our table, lets fetch the table. This is currently a 3 step procedure but will be simplified in `cloudos` v0.3.0.

```r
cb_apply_query(cohort, column_ids = c(10,110:119), keep_columns = F)
cohort <- cb_load_cohort(cohort@id)
df <- cb_get_participants_table(cohort = cohort, page_size = cb_participant_count(cohort)$total)
head(df)
```

Now lets format this into a phenofile

```r
df <- as.data.frame(apply(df, 2, unlist), stringsAsFactors = F)

# everything is now a string :/

df$FID <- df$EID
df <- df %>%
    rename(sex = `Participant phenotypic sex`, IID = EID) %>%
    select(c("FID", "IID", "sex"), everything())

# cast the PC columns to numeric type
df[,-1:-3] <- apply(df[,-1:-3],2,as.numeric)

# recode the sex column as 1/0
df$sex <- recode(df$sex, 'Male' = 1, 'Female' = 0, .default = NA_real_)

#filter out particpants missing sex or PCA values
df <- df[!is.na(df$sex),]
df <- df[!is.na(df$PC1),]
head(df)
```

That looks better. Now save the dataframe as a phenofile:

```r
write.table(df, file = "pheno.phe", sep = "\t", row.names = F, quote = F)
```
